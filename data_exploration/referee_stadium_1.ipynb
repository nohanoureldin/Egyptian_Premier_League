```python
import pandas as pd
```

```python
df = pd.read_csv(r'E:\EDA_Project\Dataset1\Combined\refandstadium_combined.csv')
```

```python
# Strip whitespace
df['Referee'] = df['Referee'].str.strip().str.title()
df['Stadium'] = df['Stadium'].str.strip().str.title()
df['Season'] = df['Season'].str.strip()
```

```python
# Flag missing instead of dropping
df['referee_known'] = df['Referee'].notnull()
df['stadium_known'] = df['Stadium'].notnull()
```

```python
# Fill missing with Unknown instead of dropping
df['Referee'] = df['Referee'].fillna('Unknown')
df['Stadium'] = df['Stadium'].fillna('Unknown')
```

```python
# Only fix clear duplicates/variants — don't map different venues to same name
stadium_map = {
    'Harras El-Hedoud Stadium': 'Haras El Hodoud Stadium',
    'Alexandra Stadium': 'Alexandria Stadium',
    'Borg El Arab Ground': 'Borg El Arab Stadium',
}
df['Stadium'] = df['Stadium'].replace(stadium_map)
```

```python
# Rename columns to match goals file style (PascalCase)
df = df.rename(columns={
    'referee_known': 'RefereeKnown',
    'stadium_known': 'StadiumKnown'
})
```

```python
# Add match count per referee — how experienced is each referee
referee_counts = df[df['Referee'] != 'Unknown']['Referee'].value_counts()
df['RefereeMatchCount'] = df['Referee'].map(referee_counts).fillna(0).astype(int)
```

```python
# Add match count per stadium
stadium_counts = df[df['Stadium'] != 'Unknown']['Stadium'].value_counts()
df['StadiumMatchCount'] = df['Stadium'].map(stadium_counts).fillna(0).astype(int)
```

```python
df = df.drop(columns=['RefereeKnown', 'StadiumKnown'])
```

```python
# Save
df.to_csv(r'E:\EDA_Project\Datasets\referee_stadium_final.csv', index=False, encoding='utf-8-sig')
print("Done! Shape:", df.shape)
```
