```python
import pandas as pd

df = pd.read_csv(r'E:\EDA_Project\Datasets\results_cleaned.csv')

# Drop redundant columns
df = df.drop(columns=[
    'season_start_year',
    'match_year',
    'match_month',
    'match_quarter',
    'match_weekday',
    'has_kickoff_time',
    'home_win',
    'away_win',
    'is_draw',
    'source_file'
])

# Fix team names to match goals file
team_fixes = {'Future FC': 'Modern Sport', 'Modern Future': 'Modern Sport'}
df['home_team'] = df['home_team'].replace(team_fixes)
df['away_team'] = df['away_team'].replace(team_fixes)

# Fix 0/1 to True/False to match goals file style
df['is_derby'] = df['is_derby'].map({1: True, 0: False})
df['is_season_complete'] = df['is_season_complete'].map({1: True, 0: False})

# ONLY rename season → Season to match goals and ref files
df = df.rename(columns={'season': 'Season'})

```

```python
df = df.drop(columns=['match_day', 'kickoff_time'])
```

```python
# Save
df.to_csv(r'E:\EDA_Project\Datasets\results_cleaned_final.csv', index=False, encoding='utf-8-sig')
print("Done! Shape:", df.shape)
print("Columns:", df.columns.tolist())
```
